```pyspark
rdd_file = spark.sparkContext.textFile("/home/jupyter/jupyter_default_dir/textfile.txt")
rdd_file.collect()
```

```pyspark
rdd_file.map(lambda x: x.split(" ")).collect()
```

```pyspark
rdd_temp = rdd_file.flatMap(lambda x: x.split(" "))
```

```pyspark
rdd_temp = rdd_temp.map(lambda x: (x,1))
rdd_temp.collect()
```

```pyspark
rdd_temp2=rdd_temp.reduceByKey(lambda x,y: x+y)
rdd_temp2.collect()
```

```pyspark
rdd_temp3 = rdd_temp2.map(lambda x : (x[1],x[0])).sortByKey()
rdd_temp3.collect()
```

```pyspark
rdd_temp.filter(lambda x : 'an' in x[0]).collect()
```
